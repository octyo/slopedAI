```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
path = "data_real.json"

with open(path, 'r') as file:
    data = json.load(file)

random_moves = np.array(data["random_moves"])
model_1_training = np.array(data["model_1_training"])
model_2_training = np.array(data["model_2_training"])
model_3_training = np.array(data["model_3_training"])

plt.plot(random_moves.mean(axis=1))
plt.plot(model_1_training.mean(axis=1))
plt.plot(model_2_training.mean(axis=1))
plt.plot(model_3_training.mean(axis=1))

print(random_moves.shape)
print(model_1_training.shape)
print(model_2_training.shape)
print(model_3_training.shape)

plt.title("Performance during training")
plt.xlabel("Generation")
plt.ylabel("Average score per generation")
plt.legend(["Random", "Model 1", "Model 2", "Model 3"])

plt.show()
```

```python
print(random_moves.std())
print(model_1_training.std())
print(model_2_training.std())
print(model_3_training.std())
```
